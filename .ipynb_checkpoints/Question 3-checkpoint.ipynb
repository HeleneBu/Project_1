Does the age of my house affect its price?

```python
import pandas as pd

import numpy as np

from matplotlib import pyplot as plt

%matplotlib inline

import seaborn as sns

df = pd.read_csv("clean_data.csv")
```

```python
df.head()
```

```python
df['age_house'].mean()
```

```python
plt.figure(figsize = (20,10))

ax = sns.scatterplot(x="age_house", y="price",
                      data=df)
```

The relationship here is not clear at all - in fact there seems to be a very slight dip in the price as a house gets older.
